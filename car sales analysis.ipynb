```python
from datetime import date
```

```python
df['Cars_Age'] = date.today().year - df['Year']
```

```python
df.head()
```

```python
labels=['Recent', 'Old', 'Very old']
bins=[0,5,10,20]
df['car_bin']=pd.cut(df['Cars_Age'], bins=bins, labels=labels ,)
```

```python
df.head()
```

```python
# Plot bar graph for car_bin distribution
df['car_bin'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribution of Cars by Age Bin')
plt.xlabel('Car Age Bin')
plt.ylabel('Number of Cars')
plt.xticks(rotation=0)
plt.show()
```

## Univariate and Multivariate Analysis
Let's perform some univariate and multivariate analysis, and modify datatypes or data as needed.

```python
# Univariate analysis: summary statistics and histograms
display(df.describe())
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
# Multivariate analysis: correlation heatmap and pairplot
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
sns.pairplot(df.select_dtypes(include=[np.number]))
plt.show()
```
